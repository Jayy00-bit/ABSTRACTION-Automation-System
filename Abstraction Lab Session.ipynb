```python
from abc import ABC,abstractmethod

class BuildingSystem(ABC):

    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass
    @abstractmethod
    def status(self):
        pass

class AirConditioningSystem(BuildingSystem):
    def start(self):
        print("The Air Conditioning System has been started:Climate Control Initiated")
    def stop(self):
        print("The Air Conditioning System has been stopped:Climate Control Deactivated")
    def status(self):
        print("The Air Conditioning System is operating efficiently at a cool 22 degrees")

class LightingSystem(BuildingSystem):
    def start(self):
        print("The Lighting System has been started:Auditorium Lights Turned on")
    def stop(self):
        print("The Lighting System has been stopped:Auditorium Lights Turned off")
    def status(self):
        print("The Lighting System are enabled at 60% brightness ")

class SecuritySystem(BuildingSystem):
    def start(self):
        print("The Security System has been started:Motion Detectors and Cameras Activated")
    def stop(self):
        print("The Security System has been stopped:Motion Detectors and Cameras deactivated")
    def status(self):
        print("The Security System has not detected any anomalies")

class FireAlarmSystem(BuildingSystem):
    def start(self):
        print("The Fire Alarm System has been started:Smoke Detectors are now active")
    def stop(self):
        print("The Fire Alarm System has been stopped:Smoke Detectors are now dormant")
    def status(self):
        print("The Fire Alarm System is now active, no smoke detected")

systems = [AirConditioningSystem(), LightingSystem(), SecuritySystem(), FireAlarmSystem()]

for system in systems:
    system.start()
    system.stop()
    system.status()
```
